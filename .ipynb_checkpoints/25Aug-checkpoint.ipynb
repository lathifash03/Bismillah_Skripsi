```python
import os
import json
import pandas as pd

# Folder tempat file JSON
folder = "./dataset" 

data = []

# Loop dari 1 sampai 1000
for i in range(1, 1001):
    filename = f"article_{i}.json"
    filepath = os.path.join(folder, filename)
    
    if os.path.exists(filepath):
        with open(filepath, "r", encoding="utf-8") as f:
            try:
                article = json.load(f)
                data.append({
                    "title": article.get("title", ""),
                    "text": article.get("text", ""),
                    "published": article.get("published", "")
                })
            except Exception as e:
                print(f"Gagal membaca {filename}: {e}")
    else:
        print(f"File tidak ditemukan: {filename}")
```

```python
# Buat DataFrame
df = pd.DataFrame(data, columns=["title", "text", "published"])

# Export ke CSV
df.to_csv("all_articles.csv", index=False, encoding="utf-8")

print("✅ Selesai! Data tersimpan di all_articles.csv")
```

```python
# 50 aja
import os
import json
import pandas as pd

# Folder tempat file JSON
folder = "./dataset" 

data = []

# Loop dari 1 sampai 1000
for i in range(1, 51):
    filename = f"article_{i}.json"
    filepath = os.path.join(folder, filename)
    
    if os.path.exists(filepath):
        with open(filepath, "r", encoding="utf-8") as f:
            try:
                article = json.load(f)
                data.append({
                    "title": article.get("title", ""),
                    "text": article.get("text", ""),
                    "published": article.get("published", "")
                })
            except Exception as e:
                print(f"Gagal membaca {filename}: {e}")
    else:
        print(f"File tidak ditemukan: {filename}")
```

```python
import pandas as pd
from openpyxl import Workbook

# Fungsi konversi ke format prosedural
def convert_to_procedural(text: str, published: str):
    points = []
    if published:  # tambahkan tanggal publikasi
        points.append(f"Artikel ini dipublikasikan pada {published}.")
    sentences = text.split(". ")
    for s in sentences:
        s = s.strip().replace("\n", " ")
        if s:
            if not s.endswith("."):
                s += "."
            points.append(s)
    return points

# 1. Baca CSV aslinya
df = pd.read_csv("all50_articles.csv")

# 2. Konversi ke prosedural
procedural_articles = []
for _, row in df.iterrows():
    procedural = convert_to_procedural(row["text"], row["published"])
    procedural_articles.append({
        "title": row["title"],
        "procedural_text": "\n".join(procedural)
    })

# 3. Simpan ke Excel
wb = Workbook()
ws = wb.active
ws.title = "Procedural Articles"

# header
ws.append(["Title", "Procedural Text"])

# data
for article in procedural_articles:
    ws.append([article["title"], article["procedural_text"]])

wb.save("procedural_articles.xlsx")
print("Selesai! File tersimpan sebagai procedural_articles.xlsx")
```

```python
#paired 1

import pandas as pd
df = pd.read_csv("all_articles.csv")

# Bersihkan teks kosong
df = df.dropna(subset=["text"]).copy()
df["word_count"] = df["text"].str.split().str.len()

# Filter panjang wajar
sample_pool = df.query("word_count >= 150 and word_count <= 1200")

# Ambil 100 sampel acak
sample = sample_pool.sample(n=100, random_state=42).reset_index(drop=True)

# Siapkan kolom template anotasi
sample = sample.rename(columns={"text":"descriptive_text"})
sample["procedural_steps"] = ""   # nanti diisi anotator (format: JSON array string)
sample["notes"] = ""

sample.to_csv("paired_to_label.csv", index=False)
```
